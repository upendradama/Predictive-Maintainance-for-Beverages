```python
from influxdb import InfluxDBClient
client = InfluxDBClient(host='localhost', port=8086)
```

```python
client.switch_database('beverage')
```

```python
results = client.query("select * from pet_sensor_readings")
```

```python
#Converting data into list of Dictionaries
results_list = next(iter(results))
```

```python
#Converting list of dictionaries into Dataframe
import pandas as pd
df = pd.DataFrame(results_list)
```
